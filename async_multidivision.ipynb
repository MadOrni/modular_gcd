```python
async def coro(taskid, q):
    
    
    for i in range(5):
        #print("starting Proc: " + str(taskid) )
        await asyncio.sleep(0.01)
        print( "Proc: "+ str(taskid)+   " -- Putting: " + str(i) )
        await q.put(i)
        #print("waiting Proc: " + str(taskid) + "\n" )
        await event.wait()
            
```

```python
async def main(loop):
    # Create a shared event
    event = asyncio.Event()
    print('event start state: {}'.format(event.is_set()))

    loop.call_later(
        2, functools.partial(set_event, event)
    )

    await asyncio.wait([coro1(event), coro2(event)])
    
    print('event end state: {}'.format(event.is_set()))
    
```

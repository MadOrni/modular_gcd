```python
def modular_buchberger(ideal, accuracy, prime_count):
    primes = []
    
    for i in range(prime_count):
        primes.append(random_prime(2^24,False,2^22))
    groebner = 1
    
    while groebner == 1:
        print('.........:::::::::: NEXT ITERATION ::::::::::::...............')
        display(Math(latex(ideal)))
        display(Math(latex(primes)))

        
        groebner = Buchberger(ideal, primes)
        prime_count += 1
        if prime_count % 3 == 0:
            accuracy *= 2
        primes = []
        for i in range(prime_count):
            primes.append(random_prime(2^24,False,2^22)) 
            
    return groebner
    
```

```python
def Buchberger(ideal, primes):
    
    G = list(ideal.gens())
    GS = []
    GS.extend(G)
    inital_loop_condition = true

    
    while G != GS or inital_loop_condition:
        inital_loop_condition = false
        GS = GS + list(set(G) - set(GS))
        #print("#####################################")
        #print("G")
        #display(Math(latex(G))) 
        #print("#####################################\n")

        #print("#####################################")
        #print("GS")
        #display(Math(latex(GS))) 
        #print("#####################################\n")
        #display(Math(latex(GS))) 
        for i in range(len(GS)-1):
            j = i+1
            
            p1 = GS[i]
            p2 = GS[j]
            
            s = create_s_polynom(p1, p2)
            h = s.reduce(G)
            #print("#####################################")
            #print("S polynom")
            #display(Math(latex(h))) 
            #print("#####################################\n")
            for prime in primes:
                
                
                P.<x ,y ,z > = PolynomialRing ( FiniteField(prime), order ='degrevlex')
                #temp_Ideal = ideal([str(pol) for pol in G]) 
                #temp_Ideal = ideal(x^4 - y - z, x*y*z - 1)

                #dont change ideal but G instead
                #print(type(ideal))
                
                Mod_G = P.ideal([str(pol) for pol in G])
                
                Mod_s = P(s)
                Mod_h = Mod_s.reduce(Mod_G)
                
                print("#####################################")
                print("S polynom")
                display(Math(latex(h)))
                print("Mod S polynom")
                display(Math(latex(h))) 
                print("#####################################\n")
                print("G")
                display(Math(latex(G)))
                print("Mod G")
                display(Math(latex(Mod_G)))                 
                print("#####################################\n")
            
                if(pol_has_same_structure(h, Mod_h) is not True):
                    return 1
            

            if h != 0:
                G.append(h)
            #display(Math(latex(G))) 

    return Sequence(G)
                    
            
    
    
```

```python
def pol_has_same_structure(pol_a, pol_b):
    #check length of coefficients, 0's are also included
    if len(pol_a.coefficients()) != len(pol_b.coefficients()):
        false
    return true
```

```python
def create_s_polynom(p1, p2):
    
    P.<x ,y ,z > = PolynomialRing ( QQ , order ='degrevlex')

    
    p_lcm = lcm(p1.lm(), p2.lm())
    s = p_lcm/p1.lt()*p1 - p_lcm/p2.lt()*p2
    
    s = P(s)
    
    return s
```

```python
def Buchberger(ideal, primes):
    
    G = list(ideal.gens())
    GS = []
    GS.extend(G)
    inital_loop_condition = true

    
    while G != GS or inital_loop_condition:
        inital_loop_condition = false
        GS = GS + list(set(G) - set(GS))

        for i in range(len(GS)-1):
            j = i+1
            p1 = GS[i]
            p2 = GS[j]
            s = create_s_polynom(p1, p2)
            h = s.reduce(G)
            
            for prime in primes:
                
                P.<x ,y ,z > = PolynomialRing ( GF(prime), order ='degrevlex')
                Mod_G = parent(s).ideal([str(P(pol)) for pol in G])
                
                Mod_G = P.ideal(Mod_G) 
                Mod_s = P(s)
                
                
                Mod_h = Mod_s.reduce(Mod_G)
                
                print("#####################################")
                #print("S polynom")
                #display(Math(latex(s)))
                print("Mod S polynom")
                print(type(Mod_s))
                print(type(parent(Mod_s)))
                print(parent(Mod_s)) 
                display(Math(latex(Mod_s))) 
                print("#####################################\n")
                
                #print("G")
                #display(Math(latex(G)))
                print("Mod G")
                display(Math(latex(Mod_G)))
                print(type(Mod_G))    
                print("#####################################\n")
                
                print("H")
                display(Math(latex(h)))
                print("Mod H")
                print(type(Mod_h))
                print(type(parent(Mod_h)))
                print(parent(Mod_h))
                display(Math(latex(Mod_h)))                 
                print("#####################################\n")                
                
                try:
                    pol_has_same_structure(h, Mod_h)
                except:
                    return 1

            if h != 0:
                G.append(h)
            #display(Math(latex(G))) 

    return Sequence(G)
                    
```

```python
def pol_has_same_structure(pol_a, pol_b):
    
    if len(pol_a.coefficients()) == len(pol_b.coefficients()):

        return True
    else:
        print("ERR ERR ERR ERR ERR ERR ERR ERR ERR ERR ERR ERR ERR ")
        print("A")
        display(Math(latex(pol_a))) 

        print("B")
        display(Math(latex(pol_b))) 
        print("ERR ERR ERR ERR ERR ERR ERR ERR ERR ERR ERR ERR ERR ")
        raise 
```

```python
def create_s_polynom(p1, p2):
    
    P = parent(p1) 
    
    p_lcm = lcm(p1.lm(), p2.lm())
    s = p_lcm/p1.lt()*p1 - p_lcm/p2.lt()*p2
    
    s = P(s)
    
    return s
```

```python
def Buchberger(ideal):
    
    G = list(ideal.gens())
    GS = []
    GS.extend(G)
    inital_loop_condition = true

    
    while G != GS or inital_loop_condition:
        inital_loop_condition = false
        GS = GS + list(set(G) - set(GS))
        #print("#####################################")
        #print("G")
        #display(Math(latex(G))) 
        #print("#####################################\n")

        #print("#####################################")
        #print("GS")
        #display(Math(latex(GS))) 
        #print("#####################################\n")
        #display(Math(latex(GS))) 
        for i in range(len(GS)-1):
            j = i+1
            
            p1 = GS[i]
            p2 = GS[j]

            h = create_s_polynom(p1, p2).reduce(G)
            #print("#####################################")
            #print("S polynom")
            #display(Math(latex(h))) 
            #print("#####################################\n")

            if h != 0:
                G.append(h)
            #display(Math(latex(G))) 

    return Sequence(G)
                    
            
    
    
```

```python
def create_s_polynom(p1, p2):
    
    P.<x ,y ,z > = PolynomialRing ( QQ , order ='degrevlex')

    
    p_lcm = lcm(p1.lm(), p2.lm())
    s = p_lcm/p1.lt()*p1 - p_lcm/p2.lt()*p2
    
    s = P(s)
    
    return s
```
